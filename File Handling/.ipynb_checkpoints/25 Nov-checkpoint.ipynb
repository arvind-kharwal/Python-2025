```python
import pickle
def writedata():
    with open('db1.dat','wb') as f:
        records = []
        while True:
            r = input()
            name = input()
            mark = int(input())
            data = [r,name,mark]
            records.append(data)
            pickle.dump(records,f)
            ch = input('Press y and n')
            if ch == 'n':
                break
        
writedata()
```

```python
def readdata():
    with open('db1.dat','rb') as f:
        while True:
            t = pickle.load(f)
            for i in t:
                print(i)
readdata()
```
