```python
import pickle
```

```python
def writedata():
    with open('db2.dat','wb') as f:
        records = []
        while True:
            r = int(input('Enter roll No'))
            name = input('Enter Name:')
            marks = int(input('Enter Marks'))
            data = [r,name,marks]
            records.append(data)
            ch = input('Press y for data and Press n for exit')
            if ch=='n':
                break
        pickle.dump(records,f)

writedata()
```

```python
def readdata():
    with open('db2.dat','rb') as f:
        while True:
            res = pickle.load(f)
            for i in res:
                print(i)

readdata()
```

```python
def readdata():
    try:
        with open('db2.dat','rb') as f:
            while True:
                res = pickle.load(f)
                for i in res:
                    print(i)
    except EOFError:
        f.close()        

readdata()
```
