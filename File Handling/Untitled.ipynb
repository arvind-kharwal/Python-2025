```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,8,40,6]
y1= [10,40,20,56]
x2 = [20,30,40,50]
y2 = [102,110,300,223]
plt.plot(x1,y1,'g',label='Line One', linewidth=3)
plt.plot(x2,y2,'r',label='Line Two', linewidth=3)
plt.title('First Plot')
plt.legend()
plt.grid(True,color='r')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
