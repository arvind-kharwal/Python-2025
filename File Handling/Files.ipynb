```python
with open('example.txt', 'r+') as file:
    # Read the first 5 characters
    content = file.read(5)
    print(f"Read: {content}")  # Output: Read: Hello (assuming "Hello World" in file)
    
    # Write new text starting from the current position (overwrites existing content)
    file.write("Python")
    
    # Move to the beginning to read the updated content
    file.seek(0)
    updated_content = file.read()
    print(f"Updated file content: {updated_content}")  # Output: Updated file content: HelloPython (overwrites "World")
```
