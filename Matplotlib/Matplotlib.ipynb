```python
from matplotlib import pyplot as plt
```

```python
x = [1,4,17,20]
y = [10,30,45,100]
plt.plot(x,y)
plt.title('Plot No 01')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
from matplotlib import style
style.use('grayscale')
x1 = [1,4,17,20]
y1 = [10,30,45,100]
x2 = [11,40,117,201]
y2 = [9,80,450,500]
plt.plot(x1,y1,'g',label='Line 1')
plt.plot(x2,y2,'c',label='Line 2',linewidth=8)
plt.title('Plot No 01')
plt.legend()
plt.grid(True,color='r')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
