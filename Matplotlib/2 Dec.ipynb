```python
from matplotlib import pyplot as plt
```

```python
x = [1,6,8,10,22]
y = [14,20,30,40,50]
plt.bar(x,y,label='Data 1',color='r')
plt.bar(y,x,label='Data 2',color='g')
plt.legend()
plt.title('Bar Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x = [1,6,8,10,22]
y = [14,20,30,40,50]
plt.barh(x,y,label='Data 1',color='r')
plt.barh(y,x,label='Data 2',color='g')
plt.legend()
plt.title('Bar Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
slices = [10,23,45,70,87]
names = ['CSE','ME','ECE','ELCE','IT']
cols = ['r','g','b','y','c']
plt.pie(slices,
       labels= names,
       colors = cols,
       shadow =True,
       startangle = 60,
       explode= (0,0.2,0,0.1,0),
       autopct = '%.2f%%')
plt.show()
```

```python
x = [1,5,10,15,20,25]
plt.hist(x,bins=5,edgecolor='r')
plt.title('Histogram Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x = [1,5,10,15,20,25]
b = [5,8,11,15,20]
plt.hist(x,bins=b,edgecolor='r',histtype='stepfilled')
plt.title('Histogram Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x = [1,5,10,15,20,25]
y = [10,50,6,15,23,55]
plt.scatter(x,y,c= 'r')
plt.show()
```
