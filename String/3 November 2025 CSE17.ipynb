```python
# Write a program to count vowels & consonants in a given string in python
```

```python
s = input()
vow = 'aeiouAEIOU'
v = 0
c = 0
for i in s:
    if i.isalpha():
        if i in vow:
            v=v+1
        else:
            c= c+1

print(v,c)
        
```

```python
s = input()
vow = 'aeiouAEIOU'
v = 0
c = 0
for i in s:
    if i.isalpha() and i in vow:
        v=v+1
    else:
        c= c+1

print(v,c)
        
```
