```python
s = input('Enter the string')
l = s.split()
print(l)
d = {}
for i in l:
    if i not in d:
        d[i] = 0
    d[i]=d[i]+1
print(d)
```

```python
d = {'a':2,'b':10}
print(d)
d1 = d
print(d1)
d1.update({'name':'Arvind'})
print(d1)
print(d)
```

```python
d = {'a':2,'b':10}
print(d)
d1 = d.copy()
print(d1)
d1.update({'name':'Arvind'})
print(d1)
print(d)
```

```python
import copy
d = {'a':2,'b':10}
print('d:',d)
d1 = copy.copy(d)
print('d1:',d1)
d1.update({'name':'Arvind'})
print('d1:',d1)
print('d:',d)
```

```python
import copy
d = {'a':2,'b':10}
print('d:',d)
d1 = copy.deepcopy(d)
print('d1:',d1)
d1.update({'name':'Arvind'})
print('d1:',d1)
print('d:',d)
```

```python
import copy
d = {'a':2,'b':10}
print('d:',d)
d1 = dict(d)
print('d1:',d1)
d1.update({'name':'Arvind'})
print('d1:',d1)
print('d:',d)
```
